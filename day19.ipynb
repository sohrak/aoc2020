```python
import re

def part1(rules, messages, rule='0'):
    r = rules[rule]
    
    contains_digits = True
    while contains_digits:
        contains_digits = False
        newr = ''
        parts = r.split(' ')
        for p in parts:
            if p.isnumeric():
                contains_digits = True
                newr += ' ( ' + rules[p] + ' )'
            else:
                newr += ' ' + p
        r = newr
    r = r.replace(' ', '')
    cnt = 0
    regex = re.compile(r)
    for message in messages:
        if regex.fullmatch(message):
            cnt += 1
            
    return cnt

part1(rules, messages)
```

```python
# 8: 42 | 42 8
# 11: 42 31 | 42 11 31
#
# Rewritten as regex:
# 8: (42)+
# 11: (42)+(31)+

def part2(rules, messages):
    r = rules.copy()
    r['8'] = ' 42 +'
    max_msg = max([len(m) for m in messages])
    regex = '( 42 31 )'
    for i in range(2, int(max_msg / 2)):
        regex += ' | ( ' + ('42 ' * i) + ('31 ' * i) + ')'
    r['11'] = '( ' + regex + ' )'
    return part1(r, messages)

part2(rules, messages)
```
