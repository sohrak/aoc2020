```python
def part1(lines, debug=False):
    mem = {}
    for (op, data) in lines:
        if op == 'mask':
            # applying the mask is just an and (to clear) and and or (to set)
            a = int(data.replace('X', '1'), 2)
            o = int(data.replace('X', '0'), 2)
            mask = (a, o)
            if debug:
                print(f'mask = {mask}')
        else:            
            if op not in mem:
                mem[op] = 0
            mem[op] = (int(data) & mask[0]) | mask[1]
            if debug:
                print(op, data, mem)
        
    return sum(mem.values())

part1(lines)
```

```python
# Returns list of addresses after mask is applied
def apply_mask(addr, mask):
    a = ''.join([a if m == '0' else m for (a, m) in zip(addr, mask)])
    addrs = []
    q = [a]
    while len(q) > 0:
        a = q.pop(0)
        if 'X' in a:
            q.append(a.replace('X', '0', 1))
            q.append(a.replace('X', '1', 1))
        else:
            addrs.append(a)
    return addrs
        
def part2(lines, debug=False):
    mem = {}
    for (op, data) in lines:
        if op == 'mask':
            mask = data
            if debug:
                print(f'mask = {mask}')
        else:
            for addr in apply_mask('{0:036b}'.format(int(op)), mask):
                if debug:
                    print(f'{addr} = {data}')
                mem[addr] = int(data)
                
    return sum(mem.values())

part2(lines)
```
