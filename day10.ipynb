```python
# Part 1, find differences

diff = jolts[1:] - jolts[:-1]
print(diff)

ones = np.count_nonzero(diff == 1)
threes = np.count_nonzero(diff == 3)

print(ones, threes, ones * threes)
```

```python
# Part 2, combinations

d = { i:0 for i in range(jolts.size) }
d[jolts.size - 1] = 1

for i in range(jolts.size - 2, -1, -1):
    count = 0
    for j in range(1, 4):
        if i + j < jolts.size and (jolts[i + j] - jolts[i]) <= 3:
            count += d[i + j]
    d[i] = count

print(d[0])
```
