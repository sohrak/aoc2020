```python
# part 1
def find(numbers, lookback):
    for i in range(lookback, len(numbers)):
        found = False
        pre = numbers[i - lookback:i]
        for j in range(len(pre)):
            diff = numbers[i] - pre[j]
            try:
                idx = pre.index(diff)
                if idx == j:
                    continue
                found = True
                break
            except:
                continue
        if not found:
            return numbers[i]
    return None

bad = find(numbers, lookback)
print(bad)
```

```python
# part 2
import numpy as np

nums = np.array(numbers)

def window_sum(n, w):
    x = n.cumsum()
    x[w:] = x[w:] - x[:-w]
    x[:w] = -1
    return x

for window in range(2, len(numbers)):
    x = window_sum(nums, window)
    i = np.argmax(x == bad)
    if i >= window:
        a = i - 1
        b = i - window + 1
        r = nums[i - window + 1:i]
        print(r)
        print(window, a, b, a + b)
        break
```
