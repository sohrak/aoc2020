```python
# part 1
good = 0
for (min, max, c, p) in passwords:
    cnt = p.count(c)
    if min <= cnt <= max:
        good += 1

print(good)        
```

```python
# part 2
good = 0
for (i1, i2, c, p) in passwords:
    ln = len(p)
    if i1 > ln or i2 > ln:
        print('noooo')
        continue
    c1 = p[i1 - 1] == c
    c2 = p[i2 - 1] == c
    if (c1 and not c2) or (c2 and not c1):
        good += 1
    
print(good)
```
